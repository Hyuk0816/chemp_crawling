```python
#드롭다운 메뉴 가져오는 함수 

def get_checked_radio_company(xpath):
    checked_items = []

    # 체크된 라디오 버튼 찾기
    try:
        checked_radio = driver.find_element(By.XPATH, xpath + '//input[@type="radio" and @checked]')
    

        # 라디오 버튼 바로 다음에 있는 <label> 요소 찾기
        label = driver.find_element(By.XPATH, xpath + '//input[@type="radio" and @checked]/following-sibling::label')
        label_text = label.text.strip()  # 텍스트 가져오기
        checked_items.append(label_text)
        
    except Exception as e:
        print(f"Error retrieving checked radio button text: {e}")

    return checked_items

def get_checked_checkbox_value(xpath):
    checked_items = []

    # 체크된 체크박스 찾기
    try:
        checked_checkbox = driver.find_element(By.XPATH, xpath + '//input[@type="checkbox" and @checked]')
        # 체크된 체크박스의 값 가져오기
        checked_value = checked_checkbox.get_attribute('value')
        checked_items.append(checked_value)
    except Exception as e:
        print(f"Error retrieving checked checkbox: {e}")

    return checked_items

def get_checked_checkbox_data(xpath):
    checked_items = []

    # 체크된 체크박스 찾기
    try:
        checked_checkbox = driver.find_element(By.XPATH, xpath + '//input[@type="checkbox" and @value="Y"]')
        # 체크된 체크박스의 값 가져오기
        checked_value = checked_checkbox.get_attribute('value')
        checked_items.append(checked_value)
    except Exception as e:
        print(f"Error retrieving checked checkbox: {e}")

    return checked_items


def get_dash_num(xpath):
    try:
        # 모든 input 요소 찾기
        inputs = driver.find_elements(By.XPATH, xpath + '//input[@type="text"]')
        # 각 input 요소의 value를 가져와서 리스트에 저장
        values = [input_elem.get_attribute('value') for input_elem in inputs]
        # 리스트의 값을 '-'로 결합하여 하나의 문자열로 만들기
        serial_number = ' - '.join(values)
        return serial_number
    except Exception as e:
        print(f"Error retrieving serial number: {e}")
        return None

def get_email(xpath):
    try:
        # 모든 input 요소 찾기
        inputs = driver.find_elements(By.XPATH, xpath + '//input[@type="text"]')
        # 각 input 요소의 value를 가져와서 리스트에 저장
        values = [input_elem.get_attribute('value') for input_elem in inputs]
        # 리스트의 값을 '-'로 결합하여 하나의 문자열로 만들기
        serial_number = ' @ '.join(values)
        return serial_number
    except Exception as e:
        print(f"Error retrieving serial number: {e}")
        return None
    
def get_xpath_element(xpath):
    try:
        element = driver.find_element(By.XPATH, xpath).text
        return element
    except NoSuchElementException:
        print(xpath + " notFound ")
        return None
```
