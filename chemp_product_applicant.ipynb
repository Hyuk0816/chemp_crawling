```python
#드롭다운 메뉴 가져오는 함수 
def get_selected_option(xpath):
    try:
        #select_element = driver.find_element(By.XPATH, xpath)

        # Select 객체 생성
        select = Select(xpath)

        # 선택된 옵션의 텍스트 가져오기
        selected_option = select.first_selected_option.text
        
        # 특정 문자열 제거
        if selected_option == "=선택=":
            return ""  # 또는 None, 필요에 따라 조정 가능
        
        return selected_option
    except NoSuchElementException:
        return NoSuchElementException
    
    
#체크 메뉴 가져오는 함수
def get_checked_radio(xpath):
    checked_items = []

    # 체크된 라디오 버튼 찾기
    try:
        checked_radio = driver.find_element(By.XPATH, xpath + '//input[@type="radio" and @checked]')
        # 체크된 라디오 버튼의 부모 요소에서 텍스트 가져오기
        label = checked_radio.find_element(By.XPATH, './..')  # 부모 <label> 요소 찾기
        label_text = label.text.strip()  # 텍스트 가져오기
        checked_items.append(label_text)
    except Exception as e:
        print(f"Error retrieving checked radio button text: {e}")

    return checked_items

def get_checked_checkbox_value(xpath):
    checked_items = []

    # 체크된 체크박스 찾기
    try:
        checked_checkbox = driver.find_element(By.XPATH, xpath + '//input[@type="checkbox" and @checked]')
        # 체크된 체크박스의 값 가져오기
        checked_value = checked_checkbox.get_attribute('value')
        checked_items.append(checked_value)
    except Exception as e:
        print(f"Error retrieving checked checkbox: {e}")

    return checked_items

def get_input_value(xpath):
    try:
        # 해당 XPATH의 입력 요소 찾기
        input_value = driver.find_element(By.XPATH, xpath).get_attribute('value')
        return input_value
    except Exception as e:
        print(f"Error retrieving input value: {e}")
        return None
    
def get_xpath_element(xpath):
    try:
        element = driver.find_element(By.XPATH, xpath).text
        return element
    except NoSuchElementException:
        print(xpath + " notFound ")
        return None
```
